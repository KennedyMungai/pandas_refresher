<p>Concatenating and Merging Data</p>


```python
titanic_data = sns.load_dataset('titanic')
```

```python
titanic_pclass1_data = titanic_data[titanic_data['class'] == 'First']
titanic_pclass1_data
```

```python
titanic_pclass2_data = titanic_data[titanic_data['class'] == 'Second']
titanic_pclass2_data
```

```python
titanic_pclass1_data.shape, titanic_pclass2_data.shape
```

```python
final_data = titanic_pclass1_data.append(
    titanic_pclass2_data, ignore_index=True)
final_data.shape()
```

```python
final_data = pd.concat([titanic_pclass1_data, titanic_pclass2_data])
final_data
```

<p>Concatenating Columns</p>


```python
df1 = final_data[:200]
df1.shape
```

```python
df2 = final_data[200:]
df2.shape
```

```python
final_data_2 = pd.concat([df1, df2], axis=1, ignore_index=True)
final_data_2
```

<p>Merging Data</p>


```python
scores1 = [
    {'Subject': 'Mathematics', 'Score': 85, 'Grade': 'B',
     'Remarks': 'Good', },
    {'Subject': 'History', 'Score': 98, 'Grade': 'A', 'Remarks':
        'Excellent'},
    {'Subject': 'English', 'Score': 76, 'Grade': 'C', 'Remarks':
        'Fair'},
    {'Subject': 'Chemistry', 'Score': 72, 'Grade': 'C', 'Remarks':
        'Fair'},
]

scores2 = [
    {'Subject': 'Arts', 'Score': 70, 'Grade': 'C', 'Remarks':
     'Fair'},
    {'Subject': 'Physics', 'Score': 75, 'Grade': 'C', 'Remarks':
     'Fair'},
    {'Subject': 'English', 'Score': 92, 'Grade': 'A', 'Remarks':
     'Excellent'},
    {'Subject': 'Chemistry', 'Score': 91, 'Grade': 'A', 'Remarks':
     'Excellent'},
]
```

```python
scores1_df = pd.DataFrame(scores1)
scores2_df = pd.DataFrame(scores2)
```

```python
scores1_df
```

```python
scores2_df
```

```python
join_inner_df = scores1_df.merge(scores2_df, on='Subject', how='inner')
join_inner_df
```

```python
join_inner_df = scores1_df.merge(scores2_df, on='Subject', how='left')
join_inner_df
```

```python
join_inner_df = scores1_df.merge(scores2_df, on='Subject', how='right')
join_inner_df
```

```python
join_inner_df = scores1_df.merge(scores2_df, on="Subject", how='outer')
join_inner_df
```

<p>Removing Duplicates</p>


```python
scores = [
    ['Mathematics', 85, 'Science'],
    ['English', 91, 'Arts'],
    ['History', 95, 'Chemistry'],
    ['History', 95, 'Chemistry'],
    ['English', 95, 'Chemistry'],
]
```

```python
my_df = pd.DataFrame(scores, columns=['Subject', 'Score', 'Subject'])
my_df
```

```python
result = my_df.drop_duplicates()
result
```

```python
result = my_df.drop_duplicates(keep='last')
result
```

```python
result = my_df.drop_duplicates(keep=False)
result
```

```python
result = my_df.drop_duplicates(subset=['Score'])
result
```

<p>Removing duplicate columns</p>


```python
scores = [
    ['Mathematics', 85, 'Science', 85],
    ['English', 91, 'Arts', 91],
    ['History', 95, 'Chemistry', 95],
    ['History', 95, 'Chemistry', 95],
    ['English', 95, 'Chemistry', 95],
]
```

```python
my_df = pd.DataFrame(scores, columns=['Subject', 'Score', 'Subject', 'Percentage'])
my_df
```

```python
result = my_df.loc[:, ~my_df.columns.duplicated()]
result
```

```python
result = my_df.T.drop_duplicates().T
result
```
