<h3>Data Visualization</h3>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('darkgrid')
plt.rcParams['figure.figsize'] = (10, 6)
```

```python
titanic_data = pd.read_csv('./data_sets/titanic_dataset.csv')
titanic_data.head(30)
```

<p>Plotting Histograms with Pandas</p>

```python
titanic_data['Age'].hist()
```

```python
titanic_data['Age'].plot(kind='hist')
```

```python
titanic_data['Age'].hist(bins=30)
```

```python
titanic_data.Age.hist(bins=30, color='orange')
```

<p>Pandas Line Plots</p>

```python
flights_data = sns.load_dataset('flights')
flights_data.head(40)
```

```python
flights_data.plot.line(y='passengers')
```

```python
flights_data.plot.line(y='passengers', color='orange')
```

```python
flights_data.plot.line(x='year', y='passengers', color='orange')
```

<p>Pandas Scatter Plots</p>

```python
flights_data.plot.scatter(x='year', y='passengers')
```

<p>Pandas Bar Plots</p>

```python
titanic_data.head(30)
```

```python
sex_mean = titanic_data.groupby('Sex')['Age'].mean()
sex_mean
```

```python
df = pd.DataFrame(
    {
        'Gender': ['Female', 'Male'],
        'Age': sex_mean.tolist()
    }
)

ax = df.plot.bar(x='Gender', y='Age')
```

```python
ax = df.plot.barh(x='Gender', y='Age')
```

<p>Pandas Box Plots</p>

```python
titanic_data.plot.box()
```

<p>Pandas Hexagonal Plots</p>

```python
tips_data = sns.load_dataset('tips')
tips_data.head(30)
```

```python
tips_data.plot.hexbin(x='total_bill', y='tip', gridsize=20)
```

<p>Pandas Kernel Density Plots</p>

```python
tips_data.plot.kde(y='tip', color='blue')
```

```python
tips_data.plot.kde(y='total_bill', color='red')
```
