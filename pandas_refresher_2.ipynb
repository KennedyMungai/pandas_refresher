<p>Handling Missing Categorical Values</p>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.rcParams["figure.figsize"] = (10,10)
sns.set_style("darkgrid")

titanic_data = sns.load_dataset("titanic")

titanic_data = titanic_data[['embark_town', 'age', 'fare']]
titanic_data.head()
titanic_data.isnull().mean()
```

```python
titanic_data.embark_town.value_counts().sort_values(ascending=False).plot(kind='bar')
plt.xlabel('Embark Town')
plt.ylabel('Number of Passengers')
```

```python
titanic_data.embark_town.mode()
```

```python
titanic_data.embark_town.fillna('Southampton', inplace=True)
```

```python
titanic_data.age.mode()
```

```python
titanic_data['age_mode'] = titanic_data.age.fillna(titanic_data.age.mode()[0])
titanic_data.head()
```

```python
plt.rcParams['figure.figsize'] = (10, 10)

fig = plt.figure()
ax = fig.add_subplot(111)

titanic_data['age'].plot(kind='kde', ax=ax)
titanic_data['age_mode'].plot(kind='kde', ax=ax, color='red')

lines, labels = ax.get_legend_handles_labels()
ax.legend(lines, labels, loc='best')
```

<p>Missing Category Imputation</p>

```python
plt.rcParams['figure.figsize'] = [8, 6]
sns.set_style('darkgrid')

titanic_data = sns.load_dataset('titanic')
titanic_data = titanic_data[['embark_town', 'age', 'fare']]
titanic_data.head()
titanic_data.isnull().mean()
```

```python
titanic_data.embark_town.fillna('Missing', inplace=True)
```

```python
titanic_data.embark_town.value_counts().sort_values(ascending=False).plot(kind='bar')
plt.xlabel('Embark Town')
plt.ylabel('No. Of Passengers')
```
